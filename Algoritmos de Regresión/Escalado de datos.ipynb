```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Simulamos datos de dos variables en escalas muy distintas
np.random.seed(42)
X = np.random.rand(100, 2) * [10, 1000]  # columna 1: 010, columna 2: 01000

# Escalamos con StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creamos el grfico
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Grfico antes del escalado
ax[0].scatter(X[:, 0], X[:, 1], color='royalblue')
ax[0].set_title("Antes del escalado")
ax[0].set_xlabel("Variable 1 (010)")
ax[0].set_ylabel("Variable 2 (01000)")

# Grfico despus del escalado
ax[1].scatter(X_scaled[:, 0], X_scaled[:, 1], color='seagreen')
ax[1].set_title("Despus del escalado (StandardScaler)")
ax[1].set_xlabel("Variable 1 (media0, std1)")
ax[1].set_ylabel("Variable 2 (media0, std1)")

plt.tight_layout()
plt.show()
```
